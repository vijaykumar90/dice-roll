# IMPORTING PACKAGES

```python
import random as rd
```

# GIVES RANDOM INTEGER BETWEEN 1 TO 6

```python
def roll():
    roll=rd.randint(1,6)
    return roll
```

# GIVES INDICES OF MAX NUMBERS IN AN ARRAY

```python
def max_indices(arr):
    maximum=max(arr)
    max_index=[]
    for i in range(len(arr)):
        if maximum==arr[i]:
            max_index.append(i+1)
            
    return max_index
```

# THE REAL GAME HAPPENS HERE

```python
def main_game():
    players=input('Number of players')
    rounds=input('Number of games')
    #for overall performance of players
    overall_performance={}
    
    for i in range(int(rounds)):
        arr=[]
        print('Round',i+1)
        print('---------------------------')
        
        for j in range(int(players)):
            n=roll()
            print('Player {} rolled {}'.format(j+1,n))
            arr.append(n)
            #overall_performance['Player {}'.format(j+1)]=0
            if i==0:
                overall_performance['Player {}'.format(j+1)]=0
            
        
            
        if len(arr)!=1:
            max_index=max_indices(arr)
            if len(max_index)==1:
                print('>>> Player {} won the round {}'.format(max_index[0],i+1))
                overall_performance['Player {}'.format(max_index[0])]+=1
            else:
                print('>>> The round is draw betweeen players {}'.format(max_index))
                for i in range(len(max_index)):
                    overall_performance['Player {}'.format(max_index[i])]=+1
                    
        arr=[]
        
    print('------------------------------------------')
    print('------------------------------------------')
    all_values=list(overall_performance.values())
    all_keys=list(overall_performance.keys())
    max_value=max(all_values)
    fin_arr=[]
    av=[]
    ak=[]
    
    
    for k in range(len(all_keys)):
        print('{} has won {} rounds'.format(all_keys[k],all_values[k]))
        
    print('--------------------------------------------')
    print('--------------------------------------------')
    print('--------------------------------------------')
    
    for i in range(len(all_keys)):
        if max_value==all_values[i]:
            fin_arr.append(i+1)
            
    if len(fin_arr)==1:
        print('Player {} has won the series'.format(fin_arr[0]))
    else:
        print("The series is draw between Players {}".format(fin_arr))
                  
```
